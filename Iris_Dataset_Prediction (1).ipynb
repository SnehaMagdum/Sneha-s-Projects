```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Iris-Dataset.csv")
df
```

```python
df = df.drop( columns = 'Id')
df
```

```python
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data = df)
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data= df,hue='Species')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('Species',axis=1),df['Species'],test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,predictions)
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
